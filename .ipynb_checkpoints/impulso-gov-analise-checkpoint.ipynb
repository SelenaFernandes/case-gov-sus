### Exploração de Dados em Dados de Monitoramento do Comportamento dos Usuários do Produto

Para medir o impacto de nossas soluções consideramos essencial
acompanhar o sucesso das ferramentas que disponibilizamos para nossos
usuários (gestores do SUS e profissionais da saúde).

Nesse estudo de caso pediremos para você realizar uma análise
exploratória para identificar padrões de uso e tirar conclusões sobre o
engajamento das ferramentas e atividade dos usuários que acessam a área
logada do Impulso Previne.

O exploração deve ser focada para tirar apenas conclusões sobre os 3
pontos abaixo:

* 1. Taxas de primeiro acesso na área logada

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
usuarios = pd.read_csv('usuarios_cadastrados_ip.csv')
eventos = pd.read_csv('registro_de_eventos_ip.csv')
```

```python
usuarios['criacao_data'] = pd.to_datetime(usuarios['criacao_data'])
```

```python
tipo_coluna_data = eventos['periodo_data_hora'].dtype
print("Tipo de dado da coluna 'data':", tipo_coluna_data)
```

```python
eventos['periodo_data_hora'] = pd.to_datetime(eventos['periodo_data_hora'], format='%Y%m%d%H')
```

```python
primeiros_acessos = eventos.groupby('usuario_id')['periodo_data_hora'].min()
```

Para identificarmos os primeiros acessos, basta captar o registro mais antigo de cada usuário

```python
primeiros_acessos_por_mes = primeiros_acessos.groupby(primeiros_acessos.dt.to_period("M")).count()
print(primeiros_acessos_por_mes)
```

É necessário primeiramente identificar quantos dos acessos realizados são classificados como primeiros acessos, e em seguida agrupá-los pela granularidade selecionada para uma visualização quantificada dos primeiros acessos. Nesse caso foi escolhido olhar para esses dados a nível de mês, para fins comparativos. 

```python
total_acessos_por_mes = eventos['periodo_data_hora'].dt.to_period("M").value_counts().sort_index()
print(total_acessos_por_mes)
```

Agora, é necessário contar o número total de acessos por mês, idependente se é ou não o primeiro acesso. 

```python
taxa_primeiro_acesso = primeiros_acessos_por_mes / total_acessos_por_mes
taxa_primeiro_acesso_porcentagem = taxa_primeiro_acesso * 100
taxa_primeiro_acesso_porcentagem
```

```python
plt.figure(figsize=(10, 6))
ax = taxa_primeiro_acesso_porcentagem.plot(kind='bar', color='skyblue')
plt.title('Taxa de Primeiro Acesso por Mês (%)')
plt.xlabel('Mês')
plt.ylabel('Taxa de Primeiro Acesso (%)')
plt.xticks(rotation=45)



for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}%', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

* 2. Comparação da atividade entre usuários antigos e novos

```python
dados_completos = pd.merge(eventos, usuarios, left_on='usuario_id', right_on='id_usuario')
```

```python
print(dados_completos.columns)
```

```python
dados_completos.head()
```

```python
usuarios_por_data = usuarios.groupby(usuarios['criacao_data'].dt.date).size().reset_index(name='Quantidade_Criados')
print(usuarios_por_data)
```

```python
dados_completos['status_usuario'] = dados_completos.apply(
    lambda row: 'Novo' if (row['criacao_data'].year == row['periodo_data_hora'].year and 
                           row['criacao_data'].month == row['periodo_data_hora'].month) 
                           else 'Antigo', 
    axis=1
)
```

```python
dados_completos[['criacao_data', 'periodo_data_hora', 'status_usuario']]
```

```python
eventos_por_status = dados_completos.groupby('status_usuario')['eventos'].sum()
```

```python
plt.figure(figsize=(8, 5))
eventos_por_status.plot(kind='bar', color=['lightblue', 'lightgreen'])
plt.title('Comparação do Número de Eventos: Usuários Novos vs Antigos')
plt.xlabel('Status do Usuário')
plt.ylabel('Número Total de Eventos')
plt.xticks(rotation=0)
plt.show()
```

```python
dados_completos['criacao_data'] = pd.to_datetime(dados_completos['criacao_data'])
dados_completos['periodo_data_hora'] = pd.to_datetime(dados_completos['periodo_data_hora'])
```

```python
tempo_sessao_por_status = dados_completos.groupby('status_usuario')['sessao_duracao'].sum()
```

```python
plt.figure(figsize=(8, 5))
tempo_sessao_por_status.plot(kind='bar', color=['blue', 'orange'])
plt.title('Tempo Total de Sessão: Usuários Novos vs Antigos')
plt.xlabel('Status do Usuário')
plt.ylabel('Tempo Total de Sessão (minutos)')
plt.xticks(rotation=0)

# Adicionando rótulos de dados
for p in plt.gca().patches:
    plt.gca().annotate(str(int(p.get_height())), (p.get_x() + p.get_width() / 2., p.get_height()),
                       ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

* 3. Utilização das listas nominais 

```python
urls_lista_nominal = dados_completos[dados_completos['pagina_url'].str.startswith('/lista_nominal')]['pagina_url'].unique()
urls_lista_nominal
```

```python
eventos_por_url = dados_completos[dados_completos['pagina_url'].str.startswith('/lista_nominal')].groupby('pagina_url')['eventos'].sum()
eventos_por_url
```

```python
dados_completos['periodo_data_hora'] = pd.to_datetime(dados_completos['periodo_data_hora'])
dados_lista_nominal = dados_completos[dados_completos['pagina_url'].str.startswith('/lista_nominal')]
```

```python
eventos_por_mes_e_url = dados_lista_nominal.groupby([dados_lista_nominal['periodo_data_hora'].dt.to_period('M'), 'pagina_url'])['eventos'].sum().unstack().fillna(0)
```

```python
eventos_por_mes_e_url.plot(kind='bar', figsize=(10, 6))
plt.title('Número de Eventos por Lista Nominal por Mês')
plt.xlabel('Mês')
plt.ylabel('Número de Eventos')
plt.xticks(rotation=45)
plt.show()
```

```python
tempo_sessao_por_url = dados_lista_nominal.groupby('pagina_url')['sessao_duracao'].sum()
tempo_sessao_por_url
```

```python
tempo_sessao_df = tempo_sessao_por_url.reset_index()
tempo_sessao_df.columns = ['pagina_url', 'tempo_sessao']
```

```python
plt.figure(figsize=(10, 6))
plt.bar(tempo_sessao_df['pagina_url'], tempo_sessao_df['tempo_sessao'], color='skyblue')
plt.title('Tempo de Sessão por Lista Nominal')
plt.xlabel('Página URL')
plt.ylabel('Tempo de Sessão (minutos)')
plt.xticks(rotation=45)

# Adicionando rótulos de dados
for i, row in tempo_sessao_df.iterrows():
    plt.text(i, row['tempo_sessao'], str(row['tempo_sessao']), ha='center', va='bottom')

plt.show()
```
